##### generate random numbers

```python
import random
def rand_num_gen(a,b):

  yield random.randint(a,b)
    
```

```python
value = rand_num_gen(1,10)
```

```python
print(value)
```

```python
for i in value:
  print(i)
  
```

##### generate multiples of 3

```python
def gen_mul_3(x,y):
  for i in range(x,y):
    if i % 3 == 0:
      yield i
```

```python
nums = gen_mul_3(5, 30)
```

```python
for num in nums:
  print(num)
```

##### generate square of numbers

```python
def gen_square():
  for num in range(1,101):
    if (num % 2 == 0) or (num % 3 == 0):
      yield num**2

```

```python
nums = gen_square()
```

```python
print(nums)
```

```python
for i in nums:
  print(i)
```

```python
# using generator comprehension
my_sq_nums = (n*n for n in range(1,100))
```
